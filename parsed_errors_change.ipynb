```python
import re
```

```python
def read_file_lines(name_file):
    f = open(name_file, 'r')
    lines = []
    for line in f:
        lines.append(line)
    return lines
```

```python
def read_hole_file(name_file):
    f = open(name_file, 'r')
    f_r = f.read()
    return f_r
```

```python
kg = []
for name in names:
    name0 = re.sub('Parsing file: .+?/', '', name)
    name0 = re.sub('\n', '', name0)
    kg.append(name0)
```

```python
final_trans = []
for j in kg:
    for h in trans_new:
        if j == h[0]:
            final_trans.append(h)
```

```python
all_f = ''
for gh in final_path:
    print(gh[0])
    empty_text = read_hole_file(gh[1])
    empty_text = empty_text.replace('. ', '.\n')
    empty_text = empty_text.replace('\n\n', '\n')
    parsed_text = read_file_lines(gh[2])
    words = []
    for every_line in parsed_text:
        w = re.search('[0-9]+\t(.*?)\t', every_line)
        if w is not None:
            words.append(w.group(1))
    word_len_mass = []
    lens = []
    i = 0
    skobka = 0
    new_words = []
    for word in words:
        string = ''
        for yet in new_words:
            string = string + yet + '.*?'
        if word == '(':
            word = '\('
        if word == ')':
            word = '\)'
        if word == '?':
            word = '\?'
        if word == '!':
            word = '\!'
    #print(string)
    #print(string + word)
        regex = re.compile(r'' + string + word, re.DOTALL)
        part = re.search(regex, empty_text)
        if i == 0:
            w_len = (word, 0, len(part.group(0)))
        else:
            w_len = (word, word_len_mass[-1][2]+1, len(part.group(0)))
        word_len_mass.append(w_len)
        if word == '.' or word == '!' or word == '?':
            new_words.append(word + '\n')
        else:
            new_words.append(word)
        i += 1
    lex_line = []
    for ev in parsed_text:
        num_tab = re.search('[0-9]+\t', ev)
        if num_tab is not None:
            lex_line.append(ev)
    k = 0
    str_new = ''
    for one in lex_line:
        t = re.search('(.+?\t){9}', one)
        if t is None:
            one = re.sub('\n', '\t\n', one)
        one = re.sub('\n','',one)
        new_one = one + '\t' + str(word_len_mass[k][1]) + '\t' + str(word_len_mass[k][2])
        str_new = str_new + new_one + '\n'
        k += 1
    str_new_mass = str_new.split('\n')
    ann_text = read_file_lines(gh[0])
    errors = []
    for ann_line in ann_text:
        t_num = re.search('^T[0-9]+\t', ann_line)
        if t_num is not None:
            errors.append(ann_line)
    type_place = []
    for t in errors:
        l = t.split('\t')
        type_place.append(l[1])
    err_mass = []
    for e in type_place:
        type_start_end = e.split(' ')
        err_mass.append(type_start_end)
    border = []
    for one_error in err_mass:
        error_type = one_error[0]
        start = one_error[1]
        end = one_error[2]
        find_start = re.search('[0-9]+.+?\t' + str(start) + '\t.+?\n', str_new)
        find_end = re.search('[0-9]+.+?\t' + str(end) + '\n', str_new)
        if find_start is not None and find_end is not None:
            del_n_start = re.sub('\n','\t#\t'+error_type+'\n',find_start.group(0))
            del_n_start = del_n_start.replace('|', '%%%')
            del_n_end = re.sub('\n','\t#\t'+error_type+'\n',find_end.group(0))
            del_n_end = del_n_end.replace('|', '%%%')
            border.append((del_n_start, del_n_end, error_type))
    str_new = str_new.replace('|', '%%%')
    for h in border:
        begi = re.search('(.+?)\t#', h[0]).group(1)
        begi = re.sub('\(', '&&&&&', begi)
        begi = re.sub('\)', '&&&&&&', begi)
        t_2 = re.search(begi + '\n', str_new)
        if t_2 is not None:
            str_new = re.sub(begi + '\n', h[0], str_new)
        else:
            str_new = re.sub(begi, begi + '\t#\t' + h[2], str_new)    
        endi = re.search('(.+?)\t#', h[1]).group(1)
        endi = re.sub('\(', '&&&&&', endi)
        endi = re.sub('\)', '&&&&&&', endi)
        t_3 = re.search(endi + '\n', str_new)
        if h[0] != h[1]:
            if t_3 is not None:
                str_new = re.sub(endi + '\n', h[1], str_new)
            else:
                str_new = re.sub(endi, endi + '\t#\t' + h[2], str_new) 
#res = str_new.replace('%%%','|')
    res = re.sub('\n1\t','\n\n\n1\t',str_new)
    my_file = open("result.txt", "w")
    my_file.write(res)
    my_file.close()
    f_n = open('result.txt')
    st_f = f_n.read()
    st_f = re.sub(r'\|', '%%%', st_f)
    z = 1
    for every_error in border:
        if every_error[0] != every_error[1]:
            #print('-----' + str(z) + '-----')
            st_er = re.search('([0-9]+\t.+?)#', every_error[0])#%%%
            en_er = re.search('([0-9]+\t.+?)#', every_error[1])
            error_area = re.search(st_er.group(1) + '(.*?\n)+' + en_er.group(1) + '.*?\n', st_f)
            #print(error_area)
            if error_area is not None:
                error_area_mass = error_area.group(0).split('\n')[:-1:]
                #print(error_area_mass)
                if len(error_area_mass) != 2:
                    n = ''
                    for each_line in error_area_mass:
                        if every_error[2] not in each_line:
                            each_line1 = each_line + '\t#\t' + every_error[2] + '\n'
                            n = n + each_line1
                        else:
                            each_line1 = each_line + '\n'
                            n = n + each_line1
                    #print('БЫЛО')
                    #print(error_area.group(0))
                    #print('СТАЛО')
                    #print(n + '\n')
                    st_f = re.sub(error_area.group(0), n + '\n', st_f)
            z += 1
    st_f = re.sub('\n\n', '\n', st_f)
    st_f = re.sub('(\n){11}', '\n\n', st_f)
    st_f = re.sub('\n1\t', '\n\n\n1\t', st_f)
    st_f = re.sub('790\t792\n\n\n\n1\tItaly\t', '790\t792\n18\tItaly\t', st_f)
    st_f_mass = st_f.split('\n\n\n')
    empty_text = empty_text.replace('…', '…\n')
    empty_text = empty_text.replace('!', '!\n')
    empty_text = empty_text.replace('? ', '?\n')
    empty_text_mass = empty_text.split('\n')[:-1:]
    for x in empty_text_mass:
        if x == '':
            empty_text_mass.remove(x)
    c = 0
    stroka = ''
    for result_parsed in st_f_mass:
        stroka = stroka + '# sent_' + str(c+1) + '\n' + '# text = ' + empty_text_mass[c] + '\n' + result_parsed + '\n\n\n'
        c += 1
    all_f = all_f + stroka
```
