```python
def read_hole_file(name_file):
    f = open(name_file, 'r', encoding='utf-8')
    f_r = f.read()
    return f_r
```

```python
old_new = read_hole_file('new_filenames.csv')[1:]
old_new_lst = old_new.split('\n')
old_sep_new_lst = []
for every_old_new in old_new_lst:
    old_sep_new = every_old_new.split(';')
    old_sep_new_lst.append(old_sep_new)
```
